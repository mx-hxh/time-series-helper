```python
import timeSeriesHelper as tsh
import numpy as np
import pandas as pd
```

## Demo for check_column_homo
The function checks each column of a dataframe to see if all of its columns are homogeneous (having same dtype across all elements)

```python
testdf = pd.DataFrame(
    {'col1':[1,2,'string',3,np.nan],
     'col2':[1,2,3,4,5],
     'datetime':np.array(['2022-10-15 11:00:00', '2022-10-15 11:10:00','2022-10-15 11:20:00',
                          '2022-10-15 11:30:00','2022-10-15 12:00:00']).astype('datetime64[ns]')       
    }
)
```

```python
tsh.check_column_homo(df = testdf)
```

## Demo for check_miss_timestep
This function check a dataframe with a time series column, whether it has missing time steps

```python
isMissingStep, missingSteps, fullTimeSteps = tsh.check_miss_timestep(df = testdf,
                                                                     timeCol = 'datetime',
                                                                     maxTime = max(testdf.datetime),
                                                                     minTime = min(testdf.datetime),
                                                                     interval = np.timedelta64(10, 'm'))
```

```python
print('isMissingStep: ' + str(isMissingStep))
print('missingSteps: ' + str(missingSteps))
print('fullTimeSteps: ' + str(fullTimeSteps))
```

## Demo for fill_miss_step
This function adds the missing time steps to the pandas dataframe as  used in time series analysis and fill 

```python
tsh.fill_miss_step(testdf, timeCol = 'datetime', fullTimeSpan = fullTimeSteps, fillValue = np.nan)
```
